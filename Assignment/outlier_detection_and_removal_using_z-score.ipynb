```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:\\download\\weight-height.csv')
df.head()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.histplot(df['Height'],kde = True)

plt.subplot(1,2,2)
sns.histplot(df['Weight'],kde = True)

plt.show()
```

Here we can see that Height column is kind of normaly distributed , so we can apply z-score outlier detection technique here

```python
print("Mean value of cgpa",df['Height'].mean())
print("Std value of cgpa",df['Height'].std())
print("Min value of cgpa",df['Height'].min())
print("Max value of cgpa",df['Height'].max())
```

```python
# Finding the boundary values
print("Highest allowed",df['Height'].mean() + 3*df['Height'].std())
print("Lowest allowed",df['Height'].mean() - 3*df['Height'].std())
```
